# Feature selection through stanndardization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv(r"F:\1.02.+Multiple+linear+regression.csv")
data
```

```python
data.describe()
```

# Creating multiplr linear regression
# describe dependent and independent variable

```python
x = data[['SAT','Rand 1,2,3']]
y = data['GPA']
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
x_scaled = scaler.transform(x)
```

```python
x_scaled
```

# Regression with scaled features

```python
reg = LinearRegression()
```

```python
reg.fit(x_scaled,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Creating a summary table

```python
reg_summary = pd.DataFrame([['Bias'],['SAT'],['Rand 1,2,3']], columns = ['Features'])
reg_summary['Weight'] = reg.intercept_, reg.coef_[0], reg.coef_[1]
reg_summary
```
